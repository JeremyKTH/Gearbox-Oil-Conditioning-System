# Extract Data
- **OTSV1** : 3 way valve
- **TV12**  : Inlet Temp to Gearbox
- **TV11**  : Outlet Temp from Gearbox
- **OTGT1** : Inlet Temp to Heat Exchanger (HE)
- **OTGT2** : Assume same as OTGT3
- **OTGT3** : Temp from HE to Cooling HE
- **OTGT4** : Cooling Temp into 3 way valve
- **QV11**  : Flow Rate (Liter/Min)
- **OTGL6** : Heating valve Temp
- **KVSV1** : Cooling valve Temp
- **KVGT1** : Cooling water Temp
- **KVGT2** : Temp out from cooling HE
- **OTGP1** : Pressure into HE

```python
clean_data = data[['OTSV1', 'TV12', 'TV11', 'OTGT1', 'OTGT2', 'OTGT3', 'OTGT4', 'OTGT6', 'QV11', 'OTSV1_FB_LREAL', 'KVSV1', 'KVGT1', 'KVGT2', 'OTGP1']]
#clean_data = data[['OTSV1', 'TV12', 'TV11', 'OTGT1', 'OTGT2', 'OTGT3', 'OTGT4', 'QV11', 'KVSV1', 'KVGT1', 'KVGT2']]
clean_data
```
